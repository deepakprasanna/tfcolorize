```python
cimg_reshaped = np.reshape(cimg, [3, 128, 128])
print(cimg_reshaped.shape)
f, axarr = plt.subplots(1,3)
axarr[0].imshow(cimg_reshaped[0])
axarr[1].imshow(cimg_reshaped[1])
axarr[2].imshow(cimg_reshaped[2])
```
